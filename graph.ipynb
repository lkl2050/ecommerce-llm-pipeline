```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Define components
components = [
    {"name": "Web Scraper\n(Best Buy)", "pos": (2, 7), "color": "#FF6B6B"},
    {"name": "Product Data\n(JSON Storage)", "pos": (6, 7), "color": "#4ECDC4"},
    {"name": "LLM Processor\n(Groq API)", "pos": (10, 7), "color": "#45B7D1"},
    {"name": "FastAPI\nREST Endpoints", "pos": (6, 4), "color": "#96CEB4"},
    {"name": "Client\n(Postman/Browser)", "pos": (6, 1), "color": "#FFEAA7"}
]

# Draw components
for comp in components:
    box = FancyBboxPatch((comp["pos"][0]-1, comp["pos"][1]-0.5), 2, 1,
                         boxstyle="round,pad=0.1", facecolor=comp["color"],
                         edgecolor="black", linewidth=2)
    ax.add_patch(box)
    ax.text(comp["pos"][0], comp["pos"][1], comp["name"],
            ha="center", va="center", fontsize=10, fontweight="bold")

# Draw arrows
arrows = [
    ((3, 7), (5, 7)),  # Scraper -> Data
    ((7, 7), (9, 7)),  # Data -> LLM
    ((10, 6.5), (7, 4.5)),  # LLM -> API
    ((6, 3.5), (6, 1.5)),  # API -> Client
    ((6, 2.5), (3, 6.5)),  # Client -> Scraper (trigger)
]

for start, end in arrows:
    ax.annotate("", xy=end, xytext=start,
                arrowprops=dict(arrowstyle="->", lw=2, color="darkblue"))

ax.set_xlim(0, 12)
ax.set_ylim(0, 8)
ax.set_title("E-commerce LLM Pipeline Architecture", fontsize=16, fontweight="bold")
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(14, 6))

# Data flow steps
steps = [
    "Web Scraping\n(Playwright + Firefox)",
    "Raw Product Data\n(Title, Price, Description)",
    "LLM Processing\n(Groq API)",
    "Enhanced Data\n(Summaries + Highlights)",
    "JSON Storage\n(Persistent)",
    "REST API\n(FastAPI)"
]

positions = [(1, 3), (3, 3), (5, 3), (7, 3), (9, 3), (11, 3)]
colors = ["#FF6B6B", "#FFE66D", "#4ECDC4", "#45B7D1", "#96CEB4", "#A8E6CF"]

for i, (step, pos, color) in enumerate(zip(steps, positions, colors)):
    circle = plt.Circle(pos, 0.7, color=color, alpha=0.8)
    ax.add_patch(circle)
    ax.text(pos[0], pos[1], step, ha="center", va="center",
            fontsize=9, fontweight="bold", wrap=True)

    if i < len(steps) - 1:
        next_pos = positions[i + 1]
        ax.arrow(pos[0] + 0.7, pos[1], next_pos[0] - pos[0] - 1.4, 0,
                head_width=0.1, head_length=0.1, fc='black', ec='black')

ax.set_xlim(0, 12)
ax.set_ylim(2, 4)
ax.set_title("Data Processing Pipeline", fontsize=16, fontweight="bold")
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 8))

# Technology layers
layers = [
    {"name": "Frontend/Client", "techs": ["Postman", "Jupyter", "Web Browser"], "y": 6, "color": "#FF6B6B"},
    {"name": "API Layer", "techs": ["FastAPI", "Uvicorn", "Pydantic"], "y": 5, "color": "#4ECDC4"},
    {"name": "Processing", "techs": ["Groq LLM", "HTTPX", "AsyncIO"], "y": 4, "color": "#45B7D1"},
    {"name": "Web Scraping", "techs": ["Playwright", "Firefox", "BeautifulSoup"], "y": 3, "color": "#96CEB4"},
    {"name": "Data Storage", "techs": ["JSON Files", "Pandas", "File System"], "y": 2, "color": "#FFEAA7"},
    {"name": "Infrastructure", "techs": ["Docker", "Python 3.11", "Linux"], "y": 1, "color": "#DDA0DD"}
]

for layer in layers:
    # Draw layer box
    rect = plt.Rectangle((1, layer["y"]-0.4), 8, 0.8,
                        facecolor=layer["color"], alpha=0.3, edgecolor="black")
    ax.add_patch(rect)

    # Layer name
    ax.text(0.5, layer["y"], layer["name"], fontweight="bold", va="center", ha="right")

    # Technologies
    tech_text = " | ".join(layer["techs"])
    ax.text(5, layer["y"], tech_text, va="center", ha="center", fontsize=10)

ax.set_xlim(0, 10)
ax.set_ylim(0.5, 6.5)
ax.set_title("Technology Stack", fontsize=16, fontweight="bold")
ax.axis('off')
plt.tight_layout()
plt.show()
```
